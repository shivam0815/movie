```python
movies=movies[['id', 'title' , 'genre' , 'overview', 'popularity']]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=10000, stop_words='english')
```

```python
cv
```

```python
cv
```

```python
vectorized = cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vectorized.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectorized)
```

```python
similarity
```

```python
new_data[new_data['title']=="The Godfather"]
```

```python
sorted(list(enumerate(similarity[2]))), reverse=True, Key=lambda vector:vector[1])
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector: vector[1])
for i in distance[0:5]:
    print(new_data.iloc[i[0]].title)
```

```python
def recommend(movies):
    index=new_data[new_data['title']==movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector: vector[1])
    for i in distance[0:5]:
      print(new_data.iloc[i[0]].title)
```

```python
recommend("Iron Man")
```
